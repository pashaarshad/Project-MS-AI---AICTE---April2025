```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df= pd.read_csv('Global YouTube Statistics.csv', encoding='latin1')
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
category_subscribers_mean = df.groupby('category')['subscribers'].mean().sort_values(ascending=False)
category_subscribers_mean
```

```python
category_subscribers_mean.plot(kind='bar')
plt.title('Mean Subscribers by Category')
plt.xlabel('Category')
plt.ylabel('Mean Subscribers')
plt.show()
```

```python
channl_type_counts = df['channel_type'].value_counts()
channl_type_counts
```

```python
channl_type_counts.tail().plot(kind='pie', autopct= '%1.1f%%')
plt.title('Channle Type Distribution')
plt.show()
```

```python
plt.scatter(df['subscribers'], df['video views'])
plt.title('Scatter Plot of Subscribers vs Video Views')
plt.xlabel('Subscribers')
plt.ylabel('Video Views')
plt.show()
```

```python
country_subscribers_total = df.groupby('Country')['subscribers'].sum().sort_values(ascending=False)
country_subscribers_total
```

```python
plt.figure(figsize=(15, 6))
country_subscribers_total.plot(kind='bar')
plt.title('Totol Subscribers by Country')
plt.xlabel('Country')
plt.ylabel('Total Subscribers')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.boxplot(x='channel_type', y='highest_monthly_earnings', data=df)
plt.title('Distribution of Highest Monthly Earning by Channel Type')
plt.xticks(rotation = 90)
plt.show()
```

```python
df.groupby(['category', 'Country'])['subscribers'].sum().unstack().plot(kind='bar', stacked=True)
plt.title('Subscribers Breakdown by Category and Country')
plt.xlabel('Category')
plt.ylabel('Total Subscribers')
plt.xticks(rotation= 90)
plt.legend(title = 'Country', bbox_to_anchor = (1.05, 1), loc = 'upper left')
plt.show()
```

```python
plt.scatter(df['Gross tertiary education enrollment (%)'], df['subscribers'], alpha=0.5)
plt.title('Scatter plot of subcribers vs gross tertiary education enrollment')
plt.xlabel('Gross tertiary education enrollment')
plt.ylabel('Subscribers')
plt.show()
```

```python
df.groupby('channel_type')['Unemployment rate'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Unemployment rate by Channel Type')
plt.xlabel('Channel Type')
plt.ylabel('Average Unemployment Rate')
plt.show()
```

```python
df.groupby(['created_year', 'channel_type'])['subscribers'].mean().unstack().tail().plot(kind='bar', width= 1)
plt.title('Average Subscribers by Created Year and Channel Type')
plt.xlabel('Created Year')
plt.ylabel('Average Subscribers')
plt.legend(title= 'Channel Type', bbox_to_anchor= (1.01, 1), loc = 'upper left')
plt.show()
```

```python
df.groupby(['created_year', 'created_month'])['subscribers'].sum().unstack().head().plot(kind='area', stacked=True)
plt.title('Subscribers Grwth Over Time')
plt.xlabel('Year')
plt.ylabel('Total Subscribers')
plt.legend(title= 'Month', bbox_to_anchor= (1.05,1), loc='upper left')
plt.show()
```

```python
df.groupby(['created_year', 'Country'])['highest_yearly_earnings'].mean().unstack().head().plot(kind='line', marker='.')
plt.title('Trend of Highest Yearly Earning by Created Year and Country')
plt.xlabel('Created Year')
plt.ylabel('Average Highest Yearly Earnings')
plt.legend(title= 'Country', bbox_to_anchor= (1.05, 1), loc='upper left')
plt.xticks(rotation= 45)
plt.show()
```

```python
category_highest_monthly_earning_sum = df.groupby('category')['highest_monthly_earnings'].sum().sort_values(ascending=False)
category_highest_monthly_earning_sum
```

```python
category_highest_monthly_earning_sum.head().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percevtage of Highest Monthly Earning by Category')
plt.ylabel('')
plt.show()
```

```python
channel_created_subscribes_mean = df.groupby(['channel_type', 'created_year'])['subscribers_for_last_30_days'].mean().unstack()
channel_created_subscribes_mean
```

```python
sns.heatmap(channel_created_subscribes_mean.head(), cmap= 'YlGnBu', annot=True)
plt.title('Mean Subscribers fol Last 30 Days by Channel Type and Created Year')
plt.xlabel('Created Year')
plt.ylabel('Channel Type')
plt.show()
```

```python
df.groupby('created_month')['subscribers_for_last_30_days'].mean().plot(kind= 'line', marker= 'o')
plt.title('Average Subscribers for last 30 Daya by Month')
plt.xlabel('Month')
plt.ylabel('Subscribers for Last 30 Days')
plt.show()
```
